```python
#importing libraries
import numpy as np
import sklearn
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import pandas as pd
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#importing dataset
bank = pd.read_csv('C:/Users/ABC/Desktop/Semester 3/Dana 4820/Final Project/Dataset/Bank_filtered.csv')
bank.head()
```

```python
bank.describe()
```

```python
#dropping unnecessary columns
bank.drop(['duration','contact','month','default','pdays',],axis=1,inplace=True)
```

```python
#determining null values
bank.isnull().sum()
```

```python
#transforming all the categorical variables into numeric for further analysis
le = preprocessing.LabelEncoder()
```

```python
bank.job = le.fit_transform(bank.job)
bank.marital = le.fit_transform(bank.marital)
bank.education = le.fit_transform(bank.education)
bank.housing = le.fit_transform(bank.housing)
bank.loan = le.fit_transform(bank.loan)
```

```python
bank.head()
```

```python
#splitting the data into test train
X=bank[['age', 'job', 'marital', 'education','balance','housing','loan','day','campaign','previous']]  # Features
y=bank['y']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#applying random forest
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
rfc_pred=rfc.predict(X_test)
```

```python
#accuracy calculation
print("Accuracy:",metrics.accuracy_score(y_test, rfc_pred))
```

```python
#Important features
pp=bank.drop('y', axis=1)
feature_importances = pd.DataFrame(rfc.feature_importances_,index = pp.columns,columns=['importance']).sort_values('importance',ascending=False)
feature_importances
```

```python
model.fit(X_train, y_train)
log_pred=model.predict(X_test)
```
